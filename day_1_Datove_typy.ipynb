```python
cislo1 = int(input('Zadaj prve cislo: '))
cislo2 = int(input('Zadaj druhe cislo: '))
sucin = int(cislo1) * int(cislo2)
sucet = int(cislo1) + int(cislo2)
print(f'Sucin cisiel {cislo1} a {cislo2} je {cislo1*cislo2}.')
print(f'Sucet cisiel {cislo1} a {cislo2} je {cislo1+cislo2}.')
```
